```python
#  Insurance Purchase Prediction using Logistic Regression

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python

# Step 2: Load Dataset
data=pd.read_csv('data-insurance.csv')
data.head()
```

```python

# Step 3: Define Features and Target
x=data[['age']]
y=data['bought_insurance']

# Step 4: Train-Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)

# Step 5: Train Logistic Regression Model
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python

# Step 6: Predictions
("\n Test Data:\n", x_test)
("\n Predictions:\n", model.predict(x_test))
("\n Accuracy:", model.score(x_test, y_test))
```

```python

# Step 7: Predictions and Accuracy
x_test
model.predict(pd.DataFrame(x_test, columns=['age']))
model.score(x_test,y_test)
```

```python

# Step 8: Logistic Curve Visualization
age_range = np.linspace(x.min(), x.max(), 200).reshape(-1, 1)
prob = model.predict_proba(age_range)[:, 1]

plt.scatter(x, y, color='blue', label='Data')
plt.plot(age_range, prob, color='red', label='Logistic Curve')
plt.xlabel('Age')
plt.ylabel('Bought Insurance (Probability)')
plt.title('Logistic Regression: Age vs Insurance')
plt.legend()
plt.show()

# Step 9: Interactive Prediction
n = int(input("Enter age: "))
pred = model.predict(pd.DataFrame([[n]], columns=['age']))[0]

if pred == 1:
    print(f"✅ At age {n}, customer is likely to buy insurance.")
else:
    print(f"❌ At age {n}, customer is not likely to buy insurance.")
```
